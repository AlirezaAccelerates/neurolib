```python
def normalizetimetrace_t(trace, t, dt):
    res = np.zeros(( np.around(t/dt).astype(int) ))

    t_fraction = len(trace)/len(res)

    res[0] =  trace[0]
    for t in range(1, len(res)):
        res[t] = trace[int(t*t_fraction)]

    return res
```
